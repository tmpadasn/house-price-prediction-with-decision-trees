Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_data_df = pd.read_csv('data_from_kaggle/train.csv')
```

Data Exploration

```python
columns_with_nan = [] 
numerical_columns = []
categorical_columns = []

for col in train_data_df.columns: #Find Nans, zeros, etc.
    print(f"Column: {col}")
    print(f"value counts: {train_data_df[col].value_counts()}")
    print(f"Column Type: {train_data_df[col].dtype}")
    print(f"Unique Values: {train_data_df[col].nunique()}")
    print(f"Nans: {train_data_df[col].isna().sum()}")
    if train_data_df[col].isna().sum() >= 1: columns_with_nan.append(col)
    if train_data_df[col].dtype == "int64" or train_data_df[col].dtype == "float64": numerical_columns.append(col)
    if train_data_df[col].dtype == "object": categorical_columns.append(col)
    print("----------------------")
```

```python
print(f"Columns with nans:\n {columns_with_nan}\n")
print(f"Numerical Columns: {len(numerical_columns)}, Categorical Columns: {len(categorical_columns)}")
```

Handle Numerical Columns with nans

```python
numerical_columns_with_nan = []
nan_replace_values = {}

for col in numerical_columns:
    print(f"Column: {col}")
    print(f"value counts: {train_data_df[col].value_counts()}")
    print(f"Column Type: {train_data_df[col].dtype}")
    print(f"Unique Values: {train_data_df[col].nunique()}")
    print(f"Nans: {train_data_df[col].isna().sum()}")
    if train_data_df[col].isna().sum() >= 1: numerical_columns_with_nan.append(col)
    print("----------------------")
```

```python
print(numerical_columns_with_nan)
```

```python
print(train_data_df['LotFrontage'].value_counts())
nan_replace_values["LotFrontage"] = train_data_df['LotFrontage'].median()
```

```python
print(train_data_df['MasVnrArea'].value_counts(), train_data_df['MasVnrArea'].isna().sum())
nan_replace_values["MasVnrArea"] = 0.0
```

```python
print(train_data_df['GarageYrBlt'].value_counts(), train_data_df['GarageYrBlt'].isna().sum())
nan_replace_values["GarageYrBlt"] = train_data_df['GarageYrBlt'].median()
```

Handle Categorical Columns with nans

```python
categorical_columns_with_nan = []

for col in categorical_columns:
    print(f"Column: {col}")
    print(f"value counts: {train_data_df[col].value_counts()}")
    print(f"Column Type: {train_data_df[col].dtype}")
    print(f"Unique Values: {train_data_df[col].nunique()}")
    print(f"Nans: {train_data_df[col].isna().sum()}")
    if train_data_df[col].isna().sum() >= 1: categorical_columns_with_nan.append(col)
    print("----------------------")
```

```python
print(categorical_columns_with_nan)

for col in categorical_columns_with_nan:
    print(train_data_df[col].value_counts(), train_data_df[col].isna().sum())
    print("------------------")
```

```python
nan_replace_values["Alley"]= "NA"
nan_replace_values["MasVnrType"]= "None"
nan_replace_values["BsmtQual"]= "NA"
nan_replace_values["BsmtCond"]= "NA"
nan_replace_values["BsmtExposure"]= "NA"
nan_replace_values["BsmtFinType1"]= "NA"
nan_replace_values["BsmtFinType2"]= "NA"
nan_replace_values["Electrical"]= "NA"
nan_replace_values["FireplaceQu"]= "NA"
nan_replace_values["GarageType"]= "NA"
nan_replace_values["GarageFinish"]= "NA"
nan_replace_values["GarageQual"]= "NA"
nan_replace_values["GarageCond"]= "NA"
nan_replace_values["PoolQC"]= "NA"
nan_replace_values["Fence"]= "NA"
nan_replace_values["MiscFeature"]= "NA"
```

```python
print(nan_replace_values)
```

```python
for col in columns_with_nan:
    train_data_df[col] = train_data_df[col].fillna(nan_replace_values[col])

train_data_df.info()
```
